```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
np.random.seed(666)
```

```python
'''
Sample in a single parameter x and give the sampler as its density
function p(x) a Gaussian density with mean 2 and variance 2.

Give the sampler a proposal distribution
q(x'|x) a Gaussian pdf for x' with mean x and variance 1
'''
```

```python
N = 10**4
samples = []
x_curr = 0 #initialize sampling with x=0
for i in range(N):
  x_prop = sp.stats.norm.rvs(loc=x_curr,scale=1) #get a proposal x value from the proposal distribution (gaussian mean = x_current and std =1)
  f_curr = sp.stats.norm.pdf(x_curr,loc=2,scale=np.sqrt(2)) #compare result of current x vs. proposed x in the sampling distribution (gaussian mean 2 std sqrt(2))
  f_prop = sp.stats.norm.pdf(x_prop,loc=2,scale=np.sqrt(2))
  rng = np.random.rand() #random number to test against the ratio to see if it is accepted
  if f_prop/f_curr > rng:
    x_curr = x_prop #if ratio is bigger then we want to keep the proposed value, add it to the sample list, and run it back with the current x as the proposed x
    samples.append(x_prop)
```

```python
x = np.linspace(min(samples),max(samples),1000) #for plotting the real pdf
plt.hist(samples,density=True, label="MCMC Samples") #histogram of samples, default bins for now
plt.plot(x,sp.stats.norm.pdf(x,loc=2,scale=np.sqrt(2)),color="red",label="True PDF") #real pdf
plt.legend()
plt.title("MCMC Samples of Gaussian vs. Gaussian PDF")
plt.xlabel("Sample Value")
plt.ylabel("Sample Probability")
plt.show()
```

##Figure 1##
Here we can see the histogram of the MCMC bins vs. the actual PDF for the sample gaussian. They line up pretty well.

```python
x = np.linspace(min(samples),max(samples),1000)
plt.hist(samples,density=True, label="MCMC Samples",bins=100) #coarser binning
plt.plot(x,sp.stats.norm.pdf(x,loc=2,scale=np.sqrt(2)),color="red",label="True PDF")
plt.legend()
plt.title("MCMC Samples of Gaussian vs. Gaussian PDF")
plt.xlabel("Sample Value")
plt.ylabel("Sample Probability")
plt.show()
```

##Figure 2##
Same as above, but coarser binning, showing even better agreement between actual PDF and sample density.
