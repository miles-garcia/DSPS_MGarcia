```python

import numpy as np
import pylab as pl
#!pip install astropy
import astropy
from astropy import units as u


%pylab inline
```

```python
def square(l, ax):
  ax.plot([-l, l], [-l, -l], 'k-')
  ax.plot([-l, l], [l, l], 'k-')
  ax.plot([-l, -l], [-l, l], 'k-')
  ax.plot([l, l], [-l, l], 'k-')
```

```python

def circle(l, ax):
  theta = np.linspace(0, 360, 100)
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
```

```python
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
```

```python
count = 0
N = 10000
xs = []
ys = []
for i in range(N):
  x,y = np.random.uniform(-1,1,size=2)
  xs.append(x)
  ys.append(y)
  tf = isin(x,y,1)
  if tf == True:
    count+=1
```

```python
count/N*4 #count/N = fraction of points in the circle = ratio of points inside circle to total points = ratio of area inside circle to the total area. The area of the sqaure is known (4) so if count/N = A(circ)/A(Sq), then count/N*A(sq) = A (circ)
```

```python
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
plt.scatter(xs,ys,alpha = 0.1)
```
